#Identifying the Directors with the Most Movie Productions on Netflix

```python
df_most_movie = df.copy()
df_most_movie = df_most_movie.loc[(df_most_movie['type'] == 'Movie')]
df_most_movie['director'] = df_most_movie['director'].str.split(',')
df_most_movie = df_most_movie.explode('director').reset_index()
df_most_movie['director'] = df_most_movie['director'].str.strip()
df_most_movie_group = df_most_movie.groupby('director')['title'].count()
df_most_movie_group = df_most_movie_group.to_frame()
df_most_movie_group = df_most_movie_group.sort_values(by='title', ascending=False)
df_most_movie_group.head(1)
```

#Identifying the Directors with the Most TV Show Productions on Netflix

```python
tv_show = df.copy()
tv_show = tv_show.loc[(tv_show['type']=='TV Show')]
tv_show['director'] = tv_show['director'].str.split(',')
tv_show = tv_show.explode('director').reset_index()
tv_show['director'] = tv_show['director'].str.strip()
tv_show_group = tv_show.groupby('director')['title'].count()
tv_show_group = tv_show_group.to_frame()
tv_show_group = tv_show_group.sort_values(by='title', ascending=False)
tv_show_group['rank'] = tv_show_group['title'].rank(method='dense',ascending=False).astype(int)
tv_show_group = tv_show_group.reset_index()
tv_show_group = tv_show_group[tv_show_group['rank']==1]
tv_show_group
```

#Find each year and the total numbers of content release in India on netflix. return top 5 year with highest avg content release

```python
india = df.copy()
india['country'] = india['country'].str.split(',')
india = india.explode('country')
india = india.loc[(india['country'] == 'India')]
india_group = india.groupby('release_year')['release_year'].count().reset_index(name='Count')
india_group
```

```python
india_group['rank'] = india_group['Count'].rank(method='dense', ascending=False).astype(int)
# india_group.sort_values(by='rank',ascending=True)
india_group_rank = india_group[india_group['rank'] < 6]
india_group_rank.sort_values(by='rank',ascending=True)
```
